```python
import pandas as pd
import numpy as np
```

```python
Z = np.array([[0.0846 + 0.6423j, 0.0567 + 0.3308j, 0.0573 + 0.3499j], 
             [0.0567 + 0.3308j, 0.0833 + 0.6438j, 0.0567 + 0.3308j],
             [0.0573 + 0.3499j, 0.0567 + 0.3308j, 0.0846 + 0.6432j]])
Z = Z*200
Z.shape
```

```python
alpha = np.cos(120*np.pi/180) + np.sin(120*np.pi/180)*1j
alpha
```

```python
I = np.array([[1],
              [alpha*alpha],
              [alpha]])
I *= 200
I
```

```python
def polar(x):
    print(str(np.sqrt(x.real*x.real + x.imag*x.imag)) + "/_" + str(180*np.arctan(x.imag/x.real)/np.pi))
    return (np.sqrt(x.real*x.real + x.imag*x.imag), 180*np.arctan(x.imag/x.real)/np.pi)
            
```

```python
V = Z @ I

polarV = []
for v in V:
    polarV.append(polar(v))
```

```python
polarV[1][0][0]
```

```python
S = V*I.conj()
print("As potências complexa são: \n{}".format(S))
P = S.real
print("As perdas ativas são \n{}".format(P/1000000))
```
