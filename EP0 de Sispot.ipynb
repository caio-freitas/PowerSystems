```python
import numpy as np
import pandas as pd
```

```python
# Tensão nominal da rede

VB1 = np.cdouble((138000.0/np.sqrt(3.0)) + 0j)
```

```python
# Impedâncias entre os barramentos

Z12 = np.cdouble(0.8 + 4j)
Z23 = np.cdouble(1.28 + 6.4j)
Z25 = np.cdouble(1.12 + 5.6j)
Z26 = np.cdouble(1.92 + 9.6j)
Z34 = np.cdouble(0.64 + 3.2j)
```

```python
# Potências complexas das cargas

M = 1000000

S4 = np.cdouble((20.0/3) + (0.25)*(12.0/3)*1j)*M
S3 = np.cdouble(5.0 + (10.0/3)*1j)*M
S5 = np.cdouble((40.0/3) + 10j)*M
S6 = np.cdouble((50.0/3) + (35.0/3)*1j)*M #MW

print(S3.real + S4.real + S5.real + S6.real)
```

```python
table = pd.DataFrame(columns=['V3', 'I3', 'V4', 'I4', 'V5', 'I5', 'V6', 'I6', 'V12', 'I12','V23', 'I23', 'V25', 'V26', 'V34'])


# Palpites iniciais das tensões
V3 = V4 = V5 = V6 = VB1

# Palpites iniciais das correntes

'''
A carga 3 é de Potência constante, logo temos que 
'''
I3 = (S3/V3).conj()

'''
A carga 4 é de Impedância constante, portanto
Ultilizmos a potência sob tensão nominal para calcular sua impedância
Então calculamos a corrente
'''
# I4 = (S4/V4).conj()
Z4 = (VB1*(VB1.conj()))/S4.conj()  # Z = V²/S*, constante
# Z4 = Z4/2
I4 = VB1/Z4
'''
A carga 5 é de Corrente constante, portanto
'''
I5 = (S5/V5).conj() # palpite inicial como sendo a tensão nominal da rede (dado que é de corrente constante)

I5_abs = np.absolute(I5) # Módulo da corrente I5, constante
phi5 = np.angle(VB1) - np.angle(I5) # fator de potencia de 5, constante (em radianos)
'''
A carga 6 é de Potência constante, portanto
'''
I6 = (S6/V6).conj()

#while not (I12 - I12_prev).absolute() < TOL: # loop certo
for i in range(30):                           # loop coxa, pra funcionar
    
    I12 = I3 + I4 + I5 + I6
    I23 = I3 + I4
    
    # Cálculo das perdas nas linhas

    V34 = Z34*I4
    V23 = Z23*I23
    V25 = Z25*I5
    V26 = Z26*I6
    V12 = Z12*I12
    
    # Cálculo das tensões nos barramentos

    V2 = VB1 - V12
    V3 = V2 - V23
    V4 = V3 - V34
    V5 = V2 - V25
    V6 = V2 - V26
    
    aux = pd.DataFrame(data=[[V3, I3, V4, I4, V5, I5, V6, I6, V12, I12, V23, I23, V25, V26, V34]], columns=['V3', 'I3', 'V4', 'I4', 'V5', 'I5', 'V6', 'I6', 'V12', 'I12','V23', 'I23', 'V25', 'V26', 'V34'])

    table = table.append(aux)
    
    '''
    A carga 3 é de Potência constante, logo temos que 
    '''
    I3 = (S3/V3).conj()
    '''
    A carga 4 é de Impedância constante, portanto
    '''
    I4 = V4/Z4

    '''
    A carga 5 é de Corrente constante, portanto
    '''
    I5 = np.cdouble(I5_abs*np.cos(np.angle(V5) - phi5) + I5_abs*np.sin(np.angle(V5) - phi5)*1j)
    
    '''
    A carga 6 é de Potência constante, portanto
    '''
    I6 = (S6/V6).conj()
    
table
```
